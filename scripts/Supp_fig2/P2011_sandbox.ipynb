```python
P2011 = te.loada('models/P2011/P2011.atm')
res = P2011.simulate(0,24*15,3000)

figure(figsize=(9,6))


cE34 = P2011.getValue('p25')*res['[cE4]']*res['[cE3n]']
cE34 = cE34/(P2011.getValue('p25')*res['[cLUX]']+P2011.getValue('p21')+P2011.getValue('m37')*res['[cCOP1d]']+P2011.getValue('m36')*res['[cCOP1n]'])

cGn = P2011.getValue('p28')*res['[cG]']/(P2011.getValue('p29')+P2011.getValue('m19')+P2011.getValue('p17')*res['[cE3n]'])

cEGn = P2011.getValue('p28')*res['[cEG]']+P2011.getValue('p17')*res['[cE3n]']*cGn
cEGn= cEGn/(P2011.getValue('m9')*res['[cCOP1n]']+P2011.getValue('m10')*res['[cCOP1d]']+P2011.getValue('p31'))

plot(res['time'],res['[cEC]'],'k' ,lw=3, label='EC')
plot(res['time'],res['[cE3]']+res['[cE3n]']+cEGn+cE34+res['[cEC]'],'r' ,lw=3, label='Total ELF3')
plot(res['time'],res['[cE4]']+cE34+res['[cEC]'],'b',lw=3, label='Total ELF4')
plot(res['time'],res['[cLUX]'],'orange' ,lw=3, label='LUX')


xlim(24*0,24*2)
xticks(range(0,24*2+12,12),fontsize=22)
axvline(12+5,color='black')
yticks(fontsize=22)
ylabel('A.U.', fontsize=22)
xlabel('Time (h)', fontsize=22)
legend(loc='upper right', fontsize=16)
axvspan(12,24, color='gray', alpha=0.7)
axvspan(24+12,24*2, color='gray', alpha=0.7)
title('A', fontsize=22)
#savefig('figures/EC_P2011.png', format='png', dpi=300)
```
