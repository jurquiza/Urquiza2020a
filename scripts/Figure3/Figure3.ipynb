```python
U2020_1 = te.loadAntimonyModel('U2020.1.txt')
U2020_4 = te.loadAntimonyModel('U2020.4.txt')
U2019_2_2 = te.loadAntimonyModel('U2019.2.2.txt')
U2020_1_res = U2020_1.simulate(0,24*20,1000)
U2020_4_res = U2020_4.simulate(0,24*20,1000)
U2019_2_2 = U2019_2_2.simulate(0,24*20,1000)
figure(dpi=600)
xticks(range(0,224*7,24), fontsize=22)
yticks(fontsize=22)
#yscale('log')
yticks([0,0.5,1,1.5])
plot(U2020_1_res['time']-24*10,U2020_1_res['[cP7_m]'],'--' ,lw=2.5, color='#004d99')
plot(U2019_2_2['time']-24*10,U2019_2_2['[cP7_mt]'], lw=2.5,color=(221/255.,38/255.,44/255.))
xlim(-24,24*4)
ylim(0,1.5)
tight_layout()
axvspan(-12,0, color='gray')
for i in range(0,6):
    axvspan(12+24*i,24*(i+1), color='gray', alpha=0.6)
savefig('PRR7_U2020_1_vs_Proposed_U2020_4.svg', format='svg', dpi=600, transparent=True)
```
